```python
import numpy as np
def rotation(vecs,ang,ax):
#vecs: 3x1 array
#ang: rotation angle in degrees
#ax : rotation axis

#convert degrees to radians
 rad = np.radians(ang)
 c = np.cos(rad)
 s = np.sin(rad)

#3x3 identity matrix
 R = np.eye(3)

 if ax == 1: #x axis rotation
  R[0,0] = 1
  R[1,1] = c
  R[1,2] = s
  R[2,1] = -s
  R[2,2] = c

 elif ax == 2: #y axis rotation
  R[0,0] = c
  R[0,2] = -s
  R[1,1] = 1
  R[2,0] = s
  R[2,2] = c

 elif ax == 3: #z axis rotation
  R[0,0] = c
  R[0,1] = s
  R[1,0] = -s
  R[1,1] = c
  R[2,2] = 1

 #Matrix mulp
 svec =  R @ vecs
 return svec

```

```python
def mainrot(vec,nang,nax):
  #vec = [3x1] numpy array(given coordinates)
  #nang = [nx1] array of rotation angles in degrewes
  #nax = [nx1] list of rotation matrixes

  # to avoid chabging the oriignal input make a copy
  current_vec = vec.copy()
  #number of rotations
  n = len(nang)
  # make a loop
  for i in range(n):
    angle = nang[i]
    axis = nax[i]

    current_vec = rotation(current_vec,angle,axis)
  return current_vec
```
